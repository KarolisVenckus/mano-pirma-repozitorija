```python
new_size = (img.size[0]//3, img.size[1]//2)
new_size
```

```python
small_img = img.resize(new_size)
small_img
```

```python
small_img.save('image.png')
```

```python
left_x = img.size[0] - img.size[0]//3*2
left_y = 0
right_x = img.size[0]
right_y = img.size[0]
crop_box = (left_x, left_y, right_x, right_y)
print(crop_box)
```

```python
boxed_image = img.crop(crop_box)
img
```

```python
thumb_img = img.copy()
thumb_size = 64, 64
thumb_img.thumbnail(thumb_size)
thumb_img
```

```python
img_g, img_b, img_r= img.split()
```

```python
img_r = img_r.point(lambda red: red * 0.5)
img_g = img_g.point(lambda green: green * 0.1)
saturated_img = Image.merge("RGB" (img_r, img_g, img_b))
saturated_img
```
